```python
import pandas as pd
import re
```

```python
# CSV dosyasını yükleme
df = pd.read_csv('/content/drive/MyDrive/ınternal_all.csv')

# Segment tanımları
segment_definitions = [
    [(r'\\/women\\/'), 'Women'],
    [(r'\\/men\\/'), 'Men'],
    [(r'\\/accessories\\/'), 'Accessories'],
    [(r'\\/sale\\/'), 'Sale'],
    [(r'\\/new-arrivals\\/'), 'New Arrivals'],
    [(r'\\/about\\/'), 'About Us'],
    [(r'\\/contact\\/'), 'Contact'],
]
```

```python
# Segment atama fonksiyonu
def get_segment(url):
    for segment_definition in segment_definitions:
        if re.findall(segment_definition[0], url):
            return segment_definition[1]
    return 'Other'

# Segment sütunu ekleme
df['segment'] = df['Address'].apply(lambda x: get_segment(x))
```

```python
# Eski ve yeni kategori ve segment sütunlarının eklenmesi
df['old_category'] = df['Address'].apply(lambda x: get_segment(x))
df['old_segment'] = df['old_category']
df['redirected_category'] = df['old_category']
df['redirected_segment'] = df['segment']
```

```python
# Sayım ve eşleşme sütunlarının eklenmesi
df['old_count'] = df.groupby('old_category')['old_category'].transform('count')
df['redirected_count'] = df.groupby('redirected_segment')['redirected_segment'].transform('count')
df['category_match'] = df['old_category'] == df['redirected_category']
df['segment_match'] = df['old_segment'] == df['redirected_segment']
df['depth_match'] = True  # Örnek olarak sabit True değerini verdim
df['depth_difference'] = 0  # Örnek olarak sabit 0 değerini verdim
```
